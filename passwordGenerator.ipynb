```python
#PASSWORD GENERATOR
#can generate passwords with a desired length, numbers and symbols
#can determine if the password strength(weak, medium or strong)
```

```python
import random
import string

#FUNCTION FOR STRENGTH CHECKER
def strength_checker(main_password):
    has_upper = False
    has_lower = False
    has_number = False
    has_symbol = False

    for char in main_password:
        if char.isupper():
            has_upper = True
        elif char.islower():
            has_lower = True
        elif char.isdigit():
            has_number = True
        elif char in symbols:
            has_symbol= True

    if len(main_password) < 8:
        return "Weak password"
    elif has_upper and has_lower and has_number and has_symbol:
        return "Strong password"
    else:
        return "Medium password"

#MAIN
print("--------------------PASSWORD GENERATOR---------------------")

#character collections
letters = string.ascii_letters
numbers = string.digits
symbols = string.punctuation

characters = letters + numbers + symbols

#asking user for requirements
password_length = int(input("Enter the password length: "))
minimum_numbers = int(input("Enter the desired minimum numbers: "))
minimum_symbols = int(input("Enter the desired minimum symbols: "))

if minimum_numbers + minimum_symbols > password_length:
    print("\nMinimum numbers and symbols exceed the given password length")

else:
    password = []

    #adding required numbers and symbols
    for i in range(minimum_numbers):
        password.append(random.choice(numbers))
    
    for i in range(minimum_symbols):
        password.append(random.choice(symbols))

    #calculating and filling up the remaining spaces
    remaining_length = password_length - minimum_numbers - minimum_symbols
    for i in range(remaining_length):
        password.append(random.choice(characters))

    #shuffling the password
    random.shuffle(password)

    #converting the list to a string
    new_password = "".join(password)

    print(f"\nGenerated password:{new_password}")

    #calling the function to check for strength
    final_password = strength_checker(new_password)
    print(f"Password strength: {final_password}")
```
