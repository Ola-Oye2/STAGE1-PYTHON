```python
#NUMBER GUESSING GAME
#can choose desired difficulty level
#has limited tries
```

```python
#NUMBER GUESSING GAME 
import random

counter = 0

print("----------------NUMBER GUESSING GAME------------------")
difficulty_level = input("Choose difficulty level (easy/medium/hard): ").lower()

if difficulty_level == "easy":
    print("\n----------MODE:EASY------------")
    max_number = 50
    tries = 10
elif difficulty_level == "medium":
    print("\n----------MODE:MEDIUM------------")
    max_number = 100
    tries = 7
elif difficulty_level == "hard":
    print("\n----------MODE:HARD------------")
    max_number = 200
    tries = 5
else:
    print("\nInvalid difficulty level")
    raise SystemExit

random_number = random.randint(1, max_number)

print(f"You are to guess a number between 1 - {max_number}")

while tries>0:
    print(f"\nTries left: {tries}")
    guess = int(input(f"Enter your guess: "))
    
    counter +=1

    if guess > random_number:
        print(f"{guess} is too high")
        tries-=1
    elif guess < random_number:
        print(f"{guess} is too low")
        tries-=1
    else:
        print(f"\n{guess} is corrrect")
        print(f"\nThis round took you {counter} guesses")
        break

if tries == 0:
    print("\nGAME OVER!")
    print(f"The correct number is: {random_number}")
```
