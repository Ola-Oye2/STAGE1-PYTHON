```python
#SHOPPING LIST MANAGER
#allows users to: 
# -add items and qty
#-view all items with their qty 
#-update the qty of an existing item 
#-remove an item from the list once purchased 
#-sort the list alphabetically or by qty
```

```python
shoppingManager = []

#FUNCTION TO ADD ITEMS
def addItems():
    print("\n---------ADD ITEMS-----------")
    while True:
        items = input("Enter an item: ")
        itemQuantity = int(input("Enter the quantity in figures: "))
        shoppingManager.append({"Item":items , "Quantity":itemQuantity})
        print("Item added successfully")
        choice = input("\nDo you want to continue? y/n?")
        if choice == 'y':
            continue
        else:
            break

#FUNCTION TO VIEW ITEMS
def viewItems():
    print("\n---------AVAILABLE ITEMS-----------")
    if len(shoppingManager) == 0:
        print("No item has been added to the shopping list")
    else:
        for index,items in enumerate(shoppingManager,1):
            print(f"{index}. {items['Item']} - {items['Quantity']}")

#FUNCTION TO UPDATE THE QUANTITY OF AN EXISTING ITEM
def updateItem():
    print("\n---------UPDATE QUANTITY OF AN ITEM-----------")
    if len(shoppingManager) == 0:
        print("No item has been added to the shopping list")
    else:
        print("\n------ ITEMS ------")
        for index, items in enumerate(shoppingManager, 1):
            print(f"{index}. {items['Item']} - {items['Quantity']}")
        itemNumber = int(input("Enter the item number you want to update: ")) -1
        newQuantity = int(input("Enter the new quantity: "))
        shoppingManager[itemNumber]["Quantity"] = newQuantity
        print("Quantity updated successfully")

#FUNCTION TO REMOVE ITEM FROM THE LIST ONCE IT'S PURCHASED
def removeItem():
    print("\n---------REMOVE AN ITEM-----------")
    if len(shoppingManager) == 0:
        print("No item has been added to the shopping list")
    else:
        print("\n------ ITEMS ------")
        for index, items in enumerate(shoppingManager, 1):
            print(f"{index}. {items['Item']} - {items['Quantity']}")
        itemNumber = input("Enter the item number you want to remove: ") -1
        shoppingManager.pop(itemNumber)
        print("Item removed successfully")

#FUNCTION TO SORT ITEMS
def sortItems():
    print("\n---------SORT ITEMS-----------")
    if len(shoppingManager) == 0:
        print("No item has been added to the shopping list")
    else:
        print("1. Sort alphabetically")
        print("2. Sort by quantity")

        choice = int(input("Enter your choice: "))

        if choice == 1:
            shoppingManager.sort(key=lambda x: x["Item"])
            print("\nList sorted alphabetically")
            print("------ ITEMS ------")
            for index, items in enumerate(shoppingManager, 1):
                print(f"{index}. {items['Item']} - {items['Quantity']}")
        elif choice == 2:
            shoppingManager.sort(key=lambda x: x["Quantity"])
            print("\nList sorted by quantity")
            print("------ ITEMS ------")
            for index, items in enumerate(shoppingManager, 1):
                print(f"{index}. {items['Item']} - {items['Quantity']}")
        else:
            print("Invalid choice")

#FUNCTION TO EXIT
def exit():
    print("\nYou have exited the menu")

#FUNCION FOR MAIN MENU
def mainMenu():
    userName = input("Enter your username: ").upper()
    print(f"WELCOME {userName} TO YOUR SHOPPING LIST MANAGER")
    while True:
        print("\nHere is a list of what you can do:")
        print("1. Add items to the shopping list along with their quantities")
        print("2. View all items in the shopping list with their quantities")
        print("3. Update the quantity of an existing item")
        print("4. Remove an item from the list once it's purchased")
        print("5. Sort the list alphabetically or quantity")
        print("6. Exit the list")

        choice = int(input("Please make your choice: "))

        if choice == 1:
            addItems()
        elif choice == 2:
            viewItems()
        elif choice == 3:
            updateItem()
        elif choice == 4:
            removeItem()
        elif choice == 5:
            sortItems()
        elif choice == 6:
            exit()
            break
        else:
            print("INVALID INPUT!!! INPUT A VALID NUMBER")
    

mainMenu()
```
