```python
# TO DO LIST USING PYTHON
# uses -variables -list and dictionaries -functions -conditionals and loops
# what it does: 1.Add a new task 2.View all tasks 3.Remove a task 4.Mark a task as completed 5.Exit
```

```python
#AN EMPTY LIST
todoList = []

#FUNCTION TO ADD A NEW TASK
def add_task():
    task = input("\nEnter a task")
    todoList.append({"Task":task, "Status":"Pending"})
    print("Successfully added new task")

#FUNCTION TO VIEW ALL TASKS
def view_task():
    print("\n----------------------------YOUR TO DO LIST-------------------------------")
    if len(todoList) == 0:
        print("No tasks have been added\n")
    else:
        for index,task in enumerate(todoList, 1):
            print(f"{index}: {task['Task']} - {task['Status']}")

#FUNCTION TO REMOVE A TASK
def remove_task():
    if len(todoList) == 0:
        print("\nThe to do list is empty")
    else:
        search_index = int(input("\nEnter the task number you want to remove: ")) -1
        if 0 <= search_index < len(todoList):
            removed_task = todoList.pop(search_index)
            print(f"Task removed: {removed_task['Task']}")
        else:
            print("Invalid number")

#FUNCTION TO MARK A TASK AS COMPLETED
def mark_complete():
    if len(todoList) == 0:
        print("\nThe to do list is empty")
    else:
        search_index = int(input("\nEnter the task number you want to mark as complete: ")) -1
        if 0 <= search_index < len(todoList):
            todoList[search_index]['Status'] = 'Complete'
            print(f"Task {todoList[search_index]['Task']} has been marked as complete")
        else:
            print("Invalid number")


#FUNCTION TO EXIT
def exit():
    print("\nYou have exited the menu")
            
#FUNCTION TO DISPLAY A MENU
def menu():
    while True:
        print("\n---------------------TO DO LIST MENU--------------------------")
        print("1. Add a new task")
        print("2. View all tasks")
        print("3. Remove a task")
        print("4. Mark a task as completed")
        print("5. Exit")

        choice = int(input("\nEnter a number between 1-4 on the menu to get a task done or 5 to exit: "))
        if choice == 1:
            add_task()
        elif choice == 2:
            view_task()
        elif choice == 3:
            remove_task()
        elif choice == 4:
            mark_complete()
        elif choice == 5:
            exit()
            break
        else:
            print("\nINVALID INPUT!!! Try again\n") 

menu()
```
